# Phoneme & Karaoke & Katakana

```python
from pythainlp.transliterate import transliterate, romanize

def thai_to_3_algorithms(text):
    results = {}
    
    # Phoneme
    results['phoneme'] = transliterate(text, engine="ipa")
    
    # Karaoke
    results['karaoke'] = romanize(text, engine="thai2rom")
    
    # Katakana
    results['katakana'] = transliterate(text, engine="th2ja")
    
    return results
```

# Test Output

```python
text = "สวัสดี"
output = thai_to_3_algorithms(text)
for key, value in output.items():
    print(f"{key.capitalize()}: {value}")
```

# ท่าไม้ตาย

```python
from groq import Groq
import os

client = Groq(api_key=os.getenv("GROQ_API_KEY"))

def get_kataoge_conversion(thai_text):
    system_prompt = """
You are a 'Thai-to-Katakana' phonetic transcriber for Japanese Text-to-Speech (TTS).
Your goal is to write Katakana that, when read by a Japanese voice, sounds like natural Thai speech.

### RULES:
1. **Long Vowels:** Use "ー" generously for Thai long vowels (e.g., 'มาก' -> マーク).
2. **Ending Consonants:** - Words ending in 'ก' use 'ック' (e.g., 'รัก' -> ラック).
   - Words ending in 'น/ม' use 'ン/ム' (e.g., 'กิน' -> ギン, 'กลม' -> クロム).
3. **No Choppiness:** Do not use 'ッ' between every syllable. Keep it flowing.
4. **Natural Tone:** - 'จ' sounds should be 'ジャ/ジ/ジュ'.
   - 'ว' at the end should be 'ーオ' or 'ーウ' (e.g., 'หิว' -> ヒウ).
5. **Output:** Katakana ONLY. No English, No Thai, No Explanations.
6. NEVER TRANSLATE. (e.g., 'เป็น' must be 'ペン', NOT 'デス')
7. NO JAPANESE GRAMMAR. Do not add 'desu', 'wa', or any particles.
8. SOUND-BASED ONLY: Focus on how a Thai person speaks.
9. Output ONLY Katakana. No English, no Thai, no notes.

### EXAMPLES:
- ไปไหนมา: パイ ナイ マー
- กินข้าวหรือยัง: ギン カーオ ルー ヤン
- ขอบคุณครับ: コープ クン カップ
- รักนะจุ๊บๆ: ラック ナ ジュップ ジュップ
"""

    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": thai_text}
        ],
        temperature=0.3 
    )
    
    return response.choices[0].message.content
```

```python
text = "สวัสดี"
output = get_kataoge_conversion(text)

print(output)
```
